## Identify missing/non applicable data

```python
# check for inf values
ATAC_inf = ATAC_seq.isin([np.inf, -np.inf]).sum()
print("Infinite values in ATAC_seq:")
print(ATAC_inf[ATAC_inf > 0])

# check for NaN values
ATAC_missing = ATAC_seq.isna().sum()
print("NaN in ATAC_seq:")
print(ATAC_missing[ATAC_missing > 0])
```

No NaN data, 5 infinite values in -log10 best P values

```python
# show rows of infinite -log10 p values

inf_rows = ATAC_seq[ATAC_seq['_-log10_bestPvalue'].isin([np.inf, -np.inf])]

print(inf_rows)
```

An infinite ("inf") -log10 p value would indicate an extremely high read count, a very strong enrichment over background noise and/or very low variability. (p value = 0)

Four of the five peaks (501969-501976) are found on the same chromosome (chrUn_JH584304) which is an unplaced genomic scaffold. The peaks are very close to each other and to the gene Pisd-ps3, a pseudogene for phosphatidylserine decarboxylase. The peaks display extremely high accessibility scores for all cell types. Some of said scores do not vary at all for different cell types. Conservation scores for three of the peaks are 0, one is at 0.51.

The fifth peak (212977) is located on chromosome 17 and has more variable accessibility scores for different cell types. It is close to a lot of genes and has a conservation score of 0.25.

```python
#drop peaks with infinite -log10 p values
ATAC_seq = ATAC_seq[ATAC_seq['_-log10_bestPvalue'] != np.inf]

# range of -log10 p-values
pvalues = ATAC_seq["_-log10_bestPvalue"]

min_pval = pvalues.min()
max_pval = pvalues.max()
print(f"Minimum -log10 p-value: {min_pval}")
print(f"Maximum -log10 p-value: {max_pval}")
```

### P value filtering

```python
#define threshold (p = 0.05)
threshold_pval = 1.3

# Filter ATAC-seq peaks based on logp value threshold 
ATAC_seq_cleaned = ATAC_seq[ATAC_seq["_-log10_bestPvalue"] >= threshold_pval]

# ATAC-seq peaks after thresholding
ATAC_seq_cleaned
```

## Median, Mean, SD of peaks

to simplify i created a dataframe (ATAC_scores) with only peak ID and accessibility score to calculate median etc.

```python
# Extract the peakID column 
peak_id_col = ATAC_seq_cleaned[["ImmGenATAC1219.peakID"]]

# Extract the accessibility score columns
score_cols = ATAC_seq_cleaned.loc[:, "LTHSC.34-.BM":]

# Create data frame with only scores per peak
ATAC_scores = pd.concat([peak_id_col, score_cols], axis=1)
```

### Normalisation of ATAC-peaks (log2 and zscaling)

```python
# log2-transformation (needed for outlier?)
peak_signal_log = np.log2(score_cols+1)
# z-scaling (needed?)
peak_signal_log_z = scipy.stats.zscore(peak_signal_log, axis = 1)
```

### Mean, median, ds, var

```python
# Extract the peakID column 
peak_id_col = ATAC_seq_cleaned[["ImmGenATAC1219.peakID"]]

# Extract the accessibility score columns
score_cols = ATAC_seq_cleaned.loc[:, "LTHSC.34-.BM":]

# Create data frame with only scores per peak
ATAC_scores = pd.concat([peak_id_col, score_cols], axis=1)

# calculate mean accessibility score for each peak
mean_accessibility_per_peak = ATAC_scores.mean(axis=1, numeric_only=True)

# calculate median accessibility score for each peak
median_accessibility_per_peak = ATAC_scores.median(axis=1, numeric_only=True)

# calculate standard deviation of accessibility scores for each peak
sd_accessibility_per_peak = ATAC_scores.std(axis=1, numeric_only=True)

# variance within cell type
variance_within_cell = ATAC_scores.var(axis = 0, numeric_only = True)

# variance across cell types (for each peak)
variance_across_cells = ATAC_scores.var(axis = 1, numeric_only=True)

# create summary data frame
mean_med_sd_df = pd.DataFrame({
    "Mean Accessibility": mean_accessibility_per_peak,
    "Median Accessibility": median_accessibility_per_peak,
    "Standard Deviation": sd_accessibility_per_peak,
    "Variance across cells": variance_across_cells,
})

#print(ATAC_scores)
#print(mean_accessibility_per_peak)
#print(mean_med_sd_df)
print(mean_med_sd_df)
print(variance_within_cell)

```
